```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Womens Clothing E-Commerce Reviews.csv"
data_df = pd.read_csv(file_path)

# Preprocess the text data (if needed)
data_df['Review Text'] = data_df['Review Text'].fillna('')  # Fill NaN values with empty strings

# Convert the text data into a frequency vector using CountVectorizer
vectorizer = CountVectorizer(stop_words='english')  # Remove English stop words
X = vectorizer.fit_transform(data_df['Review Text'])

# Apply PCA to reduce the dimensionality
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())


# Perform K-means clustering
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(X)

data_df['Cluster'] = kmeans.labels_
# silhouette_avg = silhouette_score(X.toarray(), kmeans.labels_)
# print("Silhouette Score:", silhouette_avg)

# Visualize the clusters using PCA
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
plt.title('K-means Clustering with PCA Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Womens Clothing E-Commerce Reviews.csv"
data_df = pd.read_csv(file_path)

# Preprocess the text data (if needed)
data_df['Review Text'] = data_df['Review Text'].fillna('')  # Fill NaN values with empty strings

# Convert the text data into a TF-IDF vector using TfidfVectorizer
vectorizer = TfidfVectorizer(input='content',encoding='utf-8', 
                                                        decode_error='ignore', 
                                                        lowercase=True,
                                                        stop_words = 'english',
                                                        tokenizer = None,
                                                        ngram_range=(1, 1), 
                                                        analyzer='word', 
                                                        max_features=10000)  # Remove English stop words


X = vectorizer.fit_transform(data_df['Review Text'])

# Apply PCA to reduce the dimensionality
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Perform K-means clustering
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(X)

feature_names = vectorizer.get_feature_names()


# Get top keywords for each cluster
cluster_keywords = get_cluster_keywords(X, kmeans.labels_, feature_names)

# Print top keywords for each cluster
for cluster_id, keywords in cluster_keywords.items():
    print(f"Cluster {cluster_id} top keywords: {', '.join(keywords)}")




data_df['Cluster'] = kmeans.labels_



# print(features)


# silhouette_avg = silhouette_score(X.toarray(), kmeans.labels_)
# print("Silhouette Score:", silhouette_avg)

# Visualize the clusters using PCA
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
plt.title('K-means Clustering with PCA Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python


def get_cluster_keywords(vecs, clusters, labels, top_n=3):
    # Initialize a dictionary to count term occurrences
    cluster_keyword_ids = {cluster_id: {} for cluster_id in set(clusters)}
    
    # Loop through the vector and cluster of each document
    for vec, cluster_id in zip(vecs, clusters):
        # Inspect non-zero elements of rows of sparse matrix
        for j in vec.nonzero()[1]:
            # Check if we have seen this keyword in this cluster before
            if j not in cluster_keyword_ids[cluster_id]:
                cluster_keyword_ids[cluster_id][j] = 0
            # Add a count
            cluster_keyword_ids[cluster_id][j] += 1

    # Cluster_keyword_ids contains ids
    # We need to map back to keywords
    # Do this with the labels param
    return {
        cluster_id: [
            labels[keyword_id]  # Map from kw id to keyword
            # Sort through our keyword_id_counts
            # Only return the top n per cluster
            for keyword_id, count in sorted(
                keyword_id_counts.items(),
                key=lambda x: x[1],  # Sort from highest count to lowest
                reverse=True
            )[:top_n]
        ] for cluster_id, keyword_id_counts in cluster_keyword_ids.items()
    }
```
